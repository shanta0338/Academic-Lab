```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
```

```python
years_of_education = np.array([5, 3, 8, 10, 3, 6, 4])
monthly_income = np.array([12, 10.5, 18, 22.5, 9, 13.5, 9]) 
```

# Create DataFrame

```python
data = pd.DataFrame({
    'Years of Education': years_of_education,
    'Monthly Income (k BDT)': monthly_income
})
```

```python
# (i) Scatter Plot
plt.figure(figsize=(8, 5))
plt.scatter(data['Years of Education'], data['Monthly Income (k BDT)'], color='blue')
plt.title('Scatter Plot: Education vs Monthly Income')
plt.xlabel('Years of Education')
plt.ylabel('Monthly Income (in thousand BDT)')
plt.grid(True)
plt.tight_layout()
plt.show()

# (ii) Correlation Calculation
correlation, _ = pearsonr(years_of_education, monthly_income)
correlation
```
