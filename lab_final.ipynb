## Question no 5

```python
def f(x):
    """ 
    -x**2+12x
    """
    return -2*x+12


x_curr = 6
step_size = 0.01

for i in range(100):
    left_x = x_curr - step_size  # for left neighbor
    right_x = x_curr + step_size  # for right neighbor
    """ 
    Here we evalue the function using f(x)
    """
    eval_x_init = f(x_curr)
    eval_left = f(left_x)
    eval_right = f(right_x)
    """
    Here we check the value of left and right neighbor
    """
    print(
        # print out the value of left and right neighbor
        f'epoch {i} left value: {eval_left} right value: {eval_right} current value: {eval_x_init}')
    if eval_left > eval_x_init:
        x_curr = left_x
    elif eval_right > eval_x_init:
        x_curr = right_x
    else:
        break
print(f'Current value: {f(x_curr)} at x: {x_curr}')
```
