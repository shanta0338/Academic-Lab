```python
import numpy as np
import seaborn as sns
```

```python
Values = np.array([45, 52, 40, 62, 55, 42, 48, 50, 47, 53, 40, 60])
```

```python
Values.sort()
```

```python
mean = Values.mean()
print(f'Mean : {mean}')
```

```python
median = np.median(Values, axis=0)
print(f'Median : {median}')
```

```python
print(f'Range: {Values[-1] - Values[0]}')
```

```python
print(f'Variance: {Values.var()}')
```

```python
print(f'Standard Deviation: {Values.std()}')
```

```python
import matplotlib.pyplot as plt
```

```python
values = [45, 52, 40, 62, 55, 42, 48, 50, 47, 53, 40, 60]
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Mothly Sales')
plt.tight_layout()
colors = ['red', 'blue', 'green', 'yellow', 'orange', 'purple', 
          'pink', 'brown', 'gray', 'cyan', 'magenta', 'black']
sns.barplot(x = range(1, 13), y = values, palette=colors, legend='auto',)
```
